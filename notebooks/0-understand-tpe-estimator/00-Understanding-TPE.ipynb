```python
from tpe import TPE
# Defino una función de prueba
def function(x):
    return (x-1)**3 + 2*(x-1)**2

# espacio de búsqueda:
a = -.5
b = 2

# Cre un estimador de parzen
my_estimator = TPE()

# optimizo la función
my_estimator.optimize(a, b, function, n0=6)
```

```python
# plot de las distribuciones y los intentos luego del muestreo inicial
my_estimator.plot_try(try_index=1)
```
