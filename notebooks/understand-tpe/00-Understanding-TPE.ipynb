# Three-Structured Parzen Estimator Toy Model

The Three-Structured Parzen Estimator (TPE) [[1]](#1) method works by building two density functions $\ell(x)$ and $g(x)$:

$$
p(x|y) = 
\begin{cases} 
\ell(x) & \ \text{if } y < y^* \\
g(x) & \ \text{if } y \geq y^*
\end{cases}
$$

Here you can see how it works with a simple function $ f(x) = (x-1)^3 + 2(x-1)^2$.



```python
from tpe import TPE

def function(x):
    # function to be optimized
    return (x-1)**3 + 2*(x-1)**2

# search space:
a = -.5
b = 2

# gamma is the quantile that defines the y* value
my_estimator = TPE(gamma = 0.33)

# n0 is the size for the initial random sampling
my_estimator.optimize(a, b, function, n0=6)
```

```python
# plot distributions after initial sampling
my_estimator.plot_try(try_index=1)
```

## See the optimization process

You can plot the different `try_indeces`  to see how to optimization evolved

```python
my_estimator.plot_try(try_index=5)
```

```python
# You can see the evolution of the minimu value found at any iteration
my_estimator.plot_evolution()
```
